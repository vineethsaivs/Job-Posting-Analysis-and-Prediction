# 3. Data Cleaning

**Explanation of Cleaning Process:**

1. Removing Rows with Missing Values:

2. Removing Irrelevant or Redundant Columns:

# 4. Adaptation

* Calculate and store the average years of experience required for each job posting based on the provided range.

```python
# creating a new column 'Average_Experience_Required' and converting the extracted numeric values to numeric, and calculate the average
df$Average_Experience_Required <- sapply(str_extract_all(df$Job.Experience.Required, "\\d+"), function(x) {
if (length(x) >= 2) {(as.numeric(x[1]) + as.numeric(x[2])) / 2}
else if (length(x) == 1) {as.numeric(x[1])}
else { NA }})
```

```python
head(df, 5)
```

* Count and record the total number of key skills mentioned in each job posting.

```python
# creating a new column 'Total_Key_Skills' by splitting on '|' and count the number of skills after splitting
df$Total_Key_Skills <- sapply(strsplit(df$Key.Skills, '\\|'), function(x) {length(x)})
```

```python
head(df, 5)
```

* Extract and record the primary key skill for each job posting.

```python
# creating a new column 'Top_Key_Skill' that extracts the first key skill from 'Key.Skills' and extracting the first skill if available
df$Top_Key_Skill <- sapply(strsplit(df$Key.Skills, '\\|'), function(x) {
if (length(x) >= 1) {x[1]}
else {NA}})
```

```python
head(df, 5)
```

* Extract and record the primary functional area for each job posting.

```python
# creating a new column 'Primary_Functional_Area' that extracts the first functional area from 'Functional.Area' and extracting the first functional area if available
df$Primary_Functional_Area <- sapply(strsplit(df$Functional.Area, ','), function(x) {
if (length(x) >= 1) {x[1]}
else {NA}})
```

# 8. Simple Linear Regression

### Simple Linear Regression Model1

- Step 1: Ran a linear regression model with the dependent variable sal and the independent variable Average_Experience_Required.

    - The next step is to run a linear regression model. This is a statistical method that fits a line to the data. The line is chosen to minimize the sum of the squared residuals, which are the distances between the data points and the line.

- Step 2: Summarized the model to get the coefficients.

    - Once the model is fit, we summarize it to get the coefficients. The coefficients are the parameters of the line equation. For example, in the model equation sal = m * Average_Experience_Required + b, the coefficients m and b are the slope and intercept of the line, respectively.

- Step 3: Created a scatter plot of the data with the independent variable on the x-axis and the dependent variable on the y-axis.

    - The next step is to create a scatter plot of the data. This is a graph that shows the relationship between the two variables. The independent variable is plotted on the x-axis and the dependent variable is plotted on the y-axis.

- Step 4: Added a regression line to the scatter plot.

    - Once the scatter plot is created, we can add a regression line to it. This is done by plotting the line equation from Step 3.

- Step 5: Labeled the axes and title the plot.

    - Finally, we label the axes and title the plot. The axes should be labeled with the names of the two variables. The title should be a brief description of the plot.


Interpretation of the plot:

- The plot shows a positive linear relationship between average experience and salary. This means that as average experience increases, salary also tends to increase. The regression line shows the average trend in the data.

- We can also use the regression line to make predictions about salary. For example, if we know that someone has 10 years of experience, we can use the regression line to predict that they will earn a salary of approximately $50,000.

```python
# Simple Linear Regression Model 1: Predicting salary using Average_Experience_Required
model1 <- lm(sal ~ Average_Experience_Required, data = df)
```

```python
# printing the summary of the model
summary(model1)
```

```python
# interpretation of Model 1
cat("Model 1 - Predicting Salary using Average Experience Required:\n")
cat("Coefficients:")
coef(model1)
```

```python
# visualizing the relationship between the predictor and the response
ggplot(df, aes(x = Average_Experience_Required, y = sal)) + geom_point() + geom_smooth(method = "lm", color = "blue") +
labs(title = "Simple Linear Regression - Average Experience vs. Salary")
```

Create a scatter plot of the residuals versus the fitted values. This is the residual plot.
